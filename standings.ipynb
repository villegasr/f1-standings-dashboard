```python
# Pull 2026 F1 driver standings from Ergast
url = "https://api.jolpi.ca/ergast/f1/2025/driverStandings.json"
response = requests.get(url)
data = response.json()

standings = data['MRData']['StandingsTable']['StandingsLists'][0]['DriverStandings']

# Print standings cleanly
print("\n🏎️  2025 F1 DRIVER STANDINGS\n")
for driver in standings[:10]:
    pos  = driver['position']
    name = driver['Driver']['familyName']
    pts  = driver['points']
    team = driver['Constructors'][0]['name']
    print(f"  P{pos}  {name:<20} {pts:>6} pts  —  {team}")
```
